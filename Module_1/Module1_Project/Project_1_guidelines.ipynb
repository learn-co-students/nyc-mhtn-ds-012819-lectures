
# Data Science Immersive Module 1 Project

Congratulations! 

You have made it past Module 1! You have learned so much--programming fundamentals, webscraping, API, working with Pandas, and performing exploratory analysis on datasets. You're ready to be a data analyst already. For project 1, you will collect your own data from a source of your choosing, and perform exploratory analysis and visualizations on them. 

Methodology: You are to collect data either through API calls or webscraping, put them in tabular format, and create visualizations and exploratory analysis to tell a story of these data you have collected. You and your partner are expected to use git as a collaborative tool for this project to manage version control and history.

Data Sources: anything, really. However, some interesting sources you can work with are: 
- Census/Government data:
    - Are different demographic groups different in certain areas(income,education etc?)
- Yelp
    - Are different types of cuisines rated differently?
    - Are restaurants in certain areas more likely to receive higher ratings?
- IMDB 
    - Are certain movie genres doing better in terms of box office than others?


Deliverables: Your team must prepare a science fair presentation that tells an interesting and coherent story. Your presentation should outline the process you went through and use at least 4 data visualizations to help illustrate your findings.  

Project Checklist:

- Pull data from sources of your choosing
- Establish naming conventions for variables and datasets
- Use pandas to wrangle data
- Clean dataset & record the parameters used to clean the data
- Posted to git repo:
    - A readme listing lab members, project goals, etc
    - A Jupyter notebook of clean and commented code so an independent party can replicate your analysis
    - Your final joined and cleaned dataset that was used for analysis
- A narrative Jupyter notebook that provides:
    - The purpose of your analysis and why it matters
    - 4 well annotated visualizations created using matplotlib
- Give at least two actionable insights (What type of films should they be looking to produce? What should the budget requirements be? Should they recruit certain actors for their films?)
Instructors and TCFs should be able to view a commented commit history for each element in the git repo
